```python
import numpy as np
```

```python
Scores=[[13,10,9,33],
        [63,46,90,42],
        [39,76,13,29],
        [82,9,29,78],
        [67,61,59,36]]
x=np.array(Scores)
Match_5=[41,87,72,36,92]
y=np.array(Match_5)
Match_5_2d=y.reshape(5,1)
con1=np.concatenate((Scores,Match_5_2d),axis=1)
print("Task-1 : ")
print("Adding 5th match score")
print(con1)
print("-"*25)
print("Task-2 : ")
print("Adding two new players")
Batsman_6=[77,83,98,95,89]
Batsman_7=[92,71,52,61,53]
b1=np.array(Batsman_6)
b2=np.array(Batsman_7)
b11=b1.reshape(1,5)
b21=b2.reshape(1,5)
con2=np.concatenate((con1,b11,b21),axis=0)
print(con2)
print("-"*25)
print("Task-3 :  ")
su=np.sum(con2,axis=1)
sa=np.array(su)
summ=sa.reshape(7,1)
con3=np.concatenate((con2,summ),axis=1)
print("Adding sum as column")
print(con3)
```

```python
# extract players max score
import matplotlib.pyplot as plt
x=["B-1","B-2","B-3","B-4","B-5","B-6","B-7"]
max1=np.max(con2,axis=1)
plt.bar(x,max1)
plt.show()
```

```python
# extract matchs's max score
max2=np.max(con2,axis=0)
x=["M-1","M-2","M-3","M-4","M-5"]
plt.plot(x,max2)
plt.show()
```
