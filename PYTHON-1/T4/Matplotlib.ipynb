# scatter

```python
p=["T1","T2","T3","T4"]
Jayveer=[20,24,21,20]
Dhairya=[15,16,12,9]
Pasu=[22,5,17,29]
c1=["red","blue","black","green"]
s1=[800,950,700,850]
plt.scatter(p,Jayveer,c=c1,s=s1,alpha=0.4,marker="*")
plt.scatter(p,Dhairya,c=c1,s=s1,alpha=0.2,marker="*")
plt.scatter(p,Pasu,c=c1,s=s1,alpha=0.7,marker="*")
plt.show()
```

```python
import matplotlib.pyplot as plt
m=["Comedy","Action","Romance","Drama","SciFi"]
rating=[4,5,6,1,4]

plt.subplot(5,2,1)
plt.plot(m,rating)
plt.xlabel("Movie type")
plt.ylabel("Rating")
plt.title("Line graph")

c1=["red","black","blue","Green","orange"]
s1=[300,200,250,400,350]
plt.subplot(5,2,4)
plt.scatter(m,rating,c=c1,s=s1,alpha=0.6,marker="*")
plt.title("Scatter graph")

plt.subplot(5,2,5)
plt.pie(rating,labels=m)
plt.legend(title="Survey analysis of movie")
plt.title("Pie graph")

plt.subplot(5,2,8)
plt.bar(m,rating,color="red")
plt.title("Bar graph")
plt.show()
```

# fill_between

```python
import numpy as np
import matplotlib.pyplot as plt
x1=[1,2,3,4,5,6,8]
y1=[1,1,1.5,2,2.5,4,5]
z1=[1,2,4,3,5,8,4]
plt.fill_between(x1,y1)
plt.fill_between(x1,z1,alpha=0.2)
plt.show()
```

# stackplot 

```python
import numpy as np
import matplotlib.pyplot as plt
x1=[1,2,3,4,5,6,8]
y1=[1,1,1.5,2,2.5,4,5]
z1=[1,2,4,3,5,8,4]
plt.stackplot(x1,y1,z1,labels=["A","B"])
plt.legend()
plt.show()
```
